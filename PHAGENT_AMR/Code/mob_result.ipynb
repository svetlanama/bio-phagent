```python
import pandas as pd
import glob

mob_dfs = []

files = glob.glob("/content/mob_results/barcode*/contig_report.txt")

for f in files:
    bc = f.split("barcode")[-1].split("/")[0]

    df = pd.read_csv(f, sep="\t")
    df["barcode"] = f"barcode{bc}"

    mob_dfs.append(df)

mob_all = pd.concat(mob_dfs)

mob_all
```
