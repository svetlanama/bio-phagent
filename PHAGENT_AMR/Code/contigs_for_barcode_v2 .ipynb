```python
import glob
import os
import shutil

# створюємо папки
os.makedirs("/content/results", exist_ok=True)
os.makedirs("/content/assemblies", exist_ok=True)

fastq_files = sorted(glob.glob("/content/barcode*.fastq.gz"))

for f in fastq_files:
    bc = f.split("barcode")[-1].replace(".fastq.gz", "")

    flye_out = f"/content/results/flye_barcode{bc}"
    assembly_file = f"{flye_out}/assembly.fasta"
    final_out = f"/content/assemblies/barcode{bc}_assembly.fasta"

    print(f"\n===== barcode{bc} =====")

    # 🔹 запускаємо Flye тільки якщо ще не було
    if not os.path.exists(assembly_file):
        print("Running Flye...")

        !flye --nano-raw {f} \
              --out-dir {flye_out} \
              --meta --threads 2
    else:
        print("Already assembled")

    # 🔹 копіюємо в одну папку
    if os.path.exists(assembly_file):
        shutil.copy(assembly_file, final_out)
        print(f"Saved → {final_out}")
    else:
        print("Assembly missing ❌")
```

```python
import glob
import os
import shutil

os.makedirs("/content/medaka", exist_ok=True)

assembly_files = glob.glob("/content/assemblies/barcode*_assembly.fasta")

for f in assembly_files:
    bc = f.split("barcode")[-1].replace("_assembly.fasta", "")

    fastq = f"/content/barcode{bc}.fastq.gz"
    medaka_out = f"/content/results/medaka_barcode{bc}"
    consensus_file = f"{medaka_out}/consensus.fasta"
    final_out = f"/content/medaka/barcode{bc}_consensus.fasta"

    print(f"\n===== barcode{bc} =====")

    # 🔹 запускаємо Medaka
    if not os.path.exists(consensus_file):
        print("Running Medaka...")

        !medaka_consensus -i {fastq} -d {f} -o {medaka_out} -t 2
    else:
        print("Already polished")

    # 🔹 копіюємо в одну папку
    if os.path.exists(consensus_file):
        shutil.copy(consensus_file, final_out)
        print(f"Saved → {final_out}")
    else:
        print("Consensus missing ❌")
```

```python
import glob
import pandas as pd
import os

# 🔥 ОБОВ'ЯЗКОВО
os.makedirs("/content/amr_results", exist_ok=True)

dfs = []

consensus_files = sorted(glob.glob("/content/medaka/barcode*_consensus.fasta"))

for f in consensus_files:
    bc = f.split("barcode")[-1].replace("_consensus.fasta", "")

    out_file = f"/content/amr_results/barcode{bc}_amr.tsv"

    print(f"\n===== barcode{bc} =====")
    print(f"Using: {f}")

    !amrfinder -n {f} -o {out_file}

    if os.path.exists(out_file):
        df = pd.read_csv(out_file, sep="\t")
        df["barcode"] = f"barcode{bc}"
        dfs.append(df)
        print(f"Found {len(df)} genes")
    else:
        print("No AMR output")

all_amr = pd.concat(dfs) if dfs else pd.DataFrame()

all_amr.head()
```

```python
# high-risk класи
high_priority = all_amr[
    all_amr["Class"].isin([
        "BETA-LACTAM",     # ESBL
        "CARBAPENEM",      # critical
        "COLISTIN",        # last-resort antibiotic
        "QUINOLONE"
    ])
]

high_priority[["barcode", "Element symbol", "Class"]]
```

```python
import pandas as pd
import glob

mob_dfs = []

files = glob.glob("content/mob_results/barcode*/contig_report.txt")

for f in files:
    bc = f.split("barcode")[-1].split("/")[0]

    df = pd.read_csv(f, sep="\t")
    df["barcode"] = f"barcode{bc}"

    mob_dfs.append(df)

mob_all = pd.concat(mob_dfs)

mob_all
```

```python
merged = all_amr.merge(
    mob_all,
    left_on=["barcode", "Contig id"],
    right_on=["barcode", "contig_id"],
    how="left"
)
```
